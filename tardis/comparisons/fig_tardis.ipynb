# Create the tardis comparison figure for the paper

```python
import os
os.getcwd()
```

```python
from astropy.io import ascii
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_figure(python='1d_sn.spec',tardis='/Users/long/Projects/tardis/tardis_snr.txt'):
    xpython=ascii.read(python)
    xpython.info()

    xpython['nuLnu']=xpython['Freq.']*xpython['A45P0.50']*4*np.pi*(100*3.18e18)**2


    
    xtardis=ascii.read(tardis)
    xtardis.info()
    xtardis['Lambda']=2.997e18/xtardis['nu']
    plt.plot(xpython['Lambda'],0.6e12*xpython['nuLnu'],label='Pyrite')
    plt.plot(xtardis['Lambda'],xtardis['nu']*xtardis['nuLnu'],label='Tardis')
    plt.xlim(2900,9000)
    #plt.ylim(1e52,1e55)
    plt.legend()




    
make_figure()
    
```
